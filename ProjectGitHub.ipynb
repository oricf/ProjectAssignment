```python
class Grafico(object):
    
    # Funcion principal que abre la ventana para que el usuario inicie la sesion
    def __init__(self):
        self.app = Diabetrack()
        self.util = Utilidades()
        self.principal= Tk()
        self.principal.geometry('400x650')
        self.principal.resizable(width=False,height=False)
        self.principal.title('Bienvenido')
        self.lecazucar = self.util.csv_azucar()
        self.lecinsulina = self.util.csv_insulina()
        
    # desarrollo del entorno gr'afico de la pantalla de iniciar sesion
    def inicio_sesion(self):
        self.usuariopac = Label(self.principal, text='Usuario:', font=('Courier New', 13))
        self.usuariopac.place(x=30,y=15)
        self.usuario = None
        self.introusuario = Entry(self.principal, width=35)
        self.introusuario.config(bg='#e3d3ed', bd=3)
        self.introusuario.place(x=150,y=15)
        
        self.contrasenapac = Label(self.principal, text='Contrasena:', font=('Courier New', 13))
        self.contrasenapac.place(x=30,y=115)
        self.contrasena = None
        self.introcontrasena = Entry(self.principal, width=35)
        self.introcontrasena.config(bg='#e3d3ed', bd=3)
        self.introcontrasena.place(x=150, y=115)
       
        # Cuando se presiona el boton ejecuta la funcion revisar y dira si es correcta la informacion introducida.
        self.ingresar = Button(self.principal, text='Ingresar', command=self.revisar, font=('Courier New', 13))
        self.ingresar.place(x=150, y=530)

        
        self.principal.mainloop()
        

 
    # Verificamos que el usuario y la contrasena sean las indicadas. Si estan correctas abre la ventana principal del programa,
    # si no est'a correcta abre una ventana que lo avisa.
    def revisar(self):
        
        self.usuario = self.introusuario.get()
        self.contrasena = self.introcontrasena.get()
        self.clavejunta=self.usuario + self.contrasena
        existe=self.app.comprueba_usuario(self.clavejunta)
        if not existe:
            tk.messagebox.showinfo(title='Error', message='No existe')
        else:  
            self.principal.destroy()
            self.raiz = Tk()
            self.raiz.title("Información")
            self.raiz.geometry('400x650')
            self.raiz.resizable(width=False,height=False)
    
    
            # Tab de los datos del paciente
            # Creamos un boton el cual al ser presionado abrira otra ventana, en este caso especifico abre la ventana de los datos del paciente. 
            # Tambien indicamos el estilo de la letra, las dimensiones de dicho boton y por ultimo indicamos donde los queremos localizado
            # dentro de la ventana.
            self.datos1=partial(self.datos,self.clavejunta)
            
            btnDatosPaciente = Button(self.raiz, command=self.datos1, text=("Datos del Paciente"), font=("Courier New",13), width=44, height=5).place(x=1, y=1)
            # Cargamos la imagen, le ajustamos el tamano y la posicionamos.
            fotodat = Image.open('Imagenes/DatosPacientePrincipal.png')
            resize_image = fotodat.resize((70,70))
            img_datospac = ImageTk.PhotoImage(resize_image)
            imagenpaciente = Label(self.raiz, image=img_datospac)
            imagenpaciente.place(x=25, y=15)
    
            # Tab de los medicamentos
            btnMedicamento = Button(self.raiz, command=self.medicamento1, text=("Medicamentos"), font=("Courier New",13), width=44, height=5).place(x=1, y=105)
            imagenmedic = Image.open('Imagenes/Medicamentos.png')
            resizemedic = imagenmedic.resize((70, 70))
            imagenmedic = ImageTk.PhotoImage(resizemedic)
            imagenmedicamento = Label(self.raiz, image=imagenmedic)
            imagenmedicamento.place(x=25, y=123)
    
            # Tab de Ingesta
            btnIngesta = Button(self.raiz, command=self.azucar, text=("Azúcar"), font=("Courier New",13), width=44, height=5).place(x=1, y=210)
            imagen_ingesta = Image.open('Imagenes/Ingesta.png')
            resizeingesta = imagen_ingesta.resize((70,70))
            imagen_ingesta = ImageTk.PhotoImage(resizeingesta)
            imageningesta = Label(self.raiz, image=imagen_ingesta)
            imageningesta.place(x=25, y=228)
    
    
            # Tab de Recetas
            btnCyR = Button(self.raiz, command=self.comidas, text=("Comidas y Recetas"), font=("Courier New",13), width=44, height=5).place(x=1, y=315)
            imagen_recetas = Image.open('Imagenes/Recetas.png')
            resizerecetas = imagen_recetas.resize((70,70))
            imagen_recetas = ImageTk.PhotoImage(resizerecetas)
            imagenrecetas = Label(self.raiz, image=imagen_recetas)
            imagenrecetas.place(x=25, y=333)
    
    
            # Tab de Calendario
            btnCalendario = Button(self.raiz, command=self.calendario, text=("Calendario"), font=("Courier New",13), width=44, height=5).place(x=1, y=420)
            imagen_calendario = Image.open('Imagenes/Calendario.png')
            resizecalendario = imagen_calendario.resize((70,70))
            imagen_calendario = ImageTk.PhotoImage(resizecalendario)
            imagencalendario = Label(self.raiz, image=imagen_calendario)
            imagencalendario.place(x=25, y=436)
    
    
    
            # Tab de Sintomas
            btnSintomas = Button(self.raiz, command=self.sintomas, text=("Síntomas"), font=("Courier New",13), width=44, height=5).place(x=1, y=525)
            imagen_sintomas = Image.open('Imagenes/Sintomas.png')
            resizesintomas = imagen_sintomas.resize((70,70))
            imagen_sintomas = ImageTk.PhotoImage(resizesintomas)
            imagensintomas = Label(self.raiz, image=imagen_sintomas)
            imagensintomas.place(x=25, y=541)

            # El mainloop es necesario para que se muestren las imagenes. 
            # Si no, no se pueden ver en tkinter
            self.raiz.mainloop()    
        
            
    def datos(self,clavejunta):
      
        datos_usuario=self.app.comprueba_usuario(clavejunta)
        
           
        self.datospaciente = Toplevel()
        self.datospaciente.title('Datos del Paciente')
        self.datospaciente.geometry('400x650')
        self.datospaciente.resizable(width=False,height=False)
          
         # Interfaz grafica de los datos del paciente
        nombrepac = Label(self.datospaciente, text='Nombre:', font=('Courier New', 13))
        nombrepac.place(x=30,y=15)
        nombre1 = tk.StringVar()
        nombre1.set(datos_usuario.nombre)
        intronombre = Entry(self.datospaciente, textvariable=nombre1, width=35)
        intronombre.config(bg='#e3d3ed', bd=3)
        intronombre.place(x=150,y=15)
        
        apellidospac = Label(self.datospaciente, text='Apellidos:', font=('Courier New', 13))
        apellidospac.place(x=30,y=65)
        apellidos1 = tk.StringVar()
        apellidos1.set(datos_usuario.apellidos)
        introapellidos = Entry(self.datospaciente, textvariable=apellidos1, width=35)
        introapellidos.config(bg='#e3d3ed', bd=3)
        introapellidos.place(x=150, y=65)
           
        edadpac = Label(self.datospaciente, text='Edad:', font=('Courier New', 13))
        edadpac.place(x=30,y=115)
        edad1 = tk.StringVar()
        edad1.set(datos_usuario.edad)
        introedad = Entry(self.datospaciente, textvariable=edad1, width=35)
        introedad.config(bg='#e3d3ed', bd=3)
        introedad.place(x=150, y=115)
          
        usuario = Label(self.datospaciente, text='Usuario:', font=('Courier New', 13))
        usuario.place(x=30,y=165)
        usuario1 = tk.StringVar()
        usuario1.set(datos_usuario.usuario)
        introusuario = Entry(self.datospaciente, textvariable=usuario1, width=35, state=tk.DISABLED)
        introusuario.config(bg='#e3d3ed', bd=3)
        introusuario.place(x=150, y=165)
        
        contrasena = Label(self.datospaciente, text='Contrasena:', font=('Courier New', 13))
        contrasena.place(x=30,y=215)
        contrasena1 = tk.StringVar()
        contrasena1.set(datos_usuario.contrasena)
        introcontrasena = Entry(self.datospaciente, textvariable=contrasena1, width=35, state=tk.DISABLED)
        introcontrasena.config(bg='#e3d3ed', bd=3)
        introcontrasena.place(x=150, y=215)
        
        dni = Label(self.datospaciente, text='DNI:', font=('Courier New', 13))
        dni.place(x=30,y=265)
        dni1 = tk.StringVar()
        dni1.set(datos_usuario.dni)
        introdni = Entry(self.datospaciente, textvariable=dni1, width=35)
        introdni.config(bg='#e3d3ed', bd=3)
        introdni.place(x=150, y=265) 

        fnacimiento = Label(self.datospaciente, text='Fecha de nacimiento:', font=('Courier New', 13))
        fnacimiento.place(x=30,y=315)
        fnacimiento1 = tk.StringVar()
        fnacimiento1.set(datos_usuario.fnacimiento)
        introfnacimiento = Entry(self.datospaciente, textvariable=fnacimiento1, width=18)
        introfnacimiento.config(bg='#e3d3ed', bd=3)
        introfnacimiento.place(x=250, y=315)      
```
